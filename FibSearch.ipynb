```python
import random
def busca(x,A1):
  existe=True
  for i in range(len(A1)):
    if x==A1[i]:
      existe=False
      break
  return existe
A1=[]
i=0
while i<1000:
  x=random.randint(100,10000)
  if busca(x,A1):
    A1.append(x)
    i+=1
A1.sort()
print(A1)
```
